```python
import numpy as np
```

```python
example = [[1, 1, 0.3, 1],
           [1, 0.4, 0.5, 1], 
           [1, 0.7, 0.8, 0], 
           [1, 1, 0.4, 1], 
           [1, 1, 0.5, 1], 
           [1, 0.9, 0.3, 1], 
           [1, 0.8, 0.3, 1], 
           [1, 0.7, 0.3, 1], 
           [1, 0.7, 0.7, 0], 
           [1, 0.7, 0.6, 0]]
```

```python
w = [0, 0, 0]
```

```python
perfect = False
```

```python
step = 1
```

```python
while not perfect:
    perfect = True
    for e in example:
        dot = e[0]*w[0] + e[1]*w[1] + e[2]*w[2]
        predict = int(dot > 0)
        target = e[3]
        if predict != target:
            perfect = False
            for i in range(len(w)):
                w[i] += (target-predict) * e[i]
                print( 'step ' + str(step) + ': ' + str(w) )
                step+=1
```
