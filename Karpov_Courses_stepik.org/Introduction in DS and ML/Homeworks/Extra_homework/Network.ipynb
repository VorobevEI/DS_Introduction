```python
options_number = range(10, 40, 1)
```

```python
scores_data = pd.DataFrame()
```

```python
scores_data_2 = pd.DataFrame()
```

```python
for number_neitron in options_number:
    clf_5 = MLPClassifier(hidden_layer_sizes=(number_neitron),
                      max_iter=5000, 
                      activation='logistic', 
                      learning_rate='invscaling',
                      random_state = 1,
                      warm_start = True,
                      validation_fraction = 0.3)
    clf_5.fit(X, y)
    train_score = clf_5.score(X, y)
    
    mean_cross_val_score = cross_val_score(clf_5, X, y, cv = 5, n_jobs=-1, verbose=1).mean()
    
    temp_sd = pd.DataFrame({'number_neitron': [number_neitron], 'train_score': [train_score], 'cross_val_score': [mean_cross_val_score]})
    
    scores_data_2 = scores_data_2.append(temp_sd)
```

```python
first_layer = range(5, 51, 5)
second_layer = range(5, 51, 5)
third_layer = range(5, 51, 5)
```

```python
scores_data_4 = pd.DataFrame()
```

```python
from datetime import datetime
```

```python
start_time = datetime.now()

for t in third_layer:        
    for s in second_layer:
        for f in first_layer:
            clf_5 = MLPClassifier(hidden_layer_sizes=(f, s, t),
                              max_iter=5000, 
                              activation='logistic', 
                              learning_rate='invscaling',
                              random_state = 1,
                              warm_start = True,
                              validation_fraction = 0.3)
            clf_5.fit(X, y)
            train_score = clf_5.score(X_train, y_train)
            test_score = clf_5.score(X_test, y_test)

            mean_cross_val_score = cross_val_score(clf_5, X, y, cv = 5, n_jobs=-1, verbose=1).mean()

            temp_sd = pd.DataFrame({'first_layer': [f], 'second_layer': [s], 'third_layer': [t], 'train_score': [train_score], 'test_score': [test_score], 'cross_val_score': [mean_cross_val_score]})

            scores_data_4 = scores_data_4.append(temp_sd)
            
            
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
scores_data_4.sort_values('test_score', ascending=False)
```
