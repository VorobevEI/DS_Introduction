```python
X_train = df.drop(['class_ship'], axis=1)
```

```python
clf = RandomForestClassifier(random_state=0)
```

```python
parametrs = {'n_estimators': list(range(1, 52, 5)), 
             'max_depth': list(range(1, 19, 2)),
             'min_samples_leaf': list(range(1, 11, 1)),
             'min_samples_split': list(range(1, 11, 1))}
```

```python
best_clf=GridSearchCV(clf, parametrs, cv=3, n_jobs=-1, verbose=1)
```

```python
best_clf.fit(X_train, y_train)
```

```python
best_clf.best_params_
```

```python
new_clf = RandomForestClassifier(random_state=0, n_estimators = 6, max_depth = 3, min_samples_leaf = 1, min_samples_split = 2)
```

```python
new_clf.fit(X_train, y_train)
```
