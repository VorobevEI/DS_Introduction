```python
#Importing libraries for workflow
#Importer des bibliotheques pour le flux de travail
#Импорт библиотек для работ

import pandas as pd
import numpy as np
```

```python
#Loading data frame into sp (Students Performance)
#Chargement de la dataframe dans sp
#Загрузка данных в переменную sp

sp=pd.read_csv('C:/Users/9/Downloads/StudentsPerformance.csv')
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#Renaming columns
#Renommer les colonnes
#Переименование колонок

sp = sp \
    .rename(columns = 
           {'parental level of education': 'parental_level_of_education',
            'test preparation course': 'test_preparation_course',
            'math score': 'math_score',
            'reading score': 'reading_score',
            'writing score': 'writing_score'
            })
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#Grouping by gender with mean scores
#Groupement par genre avec des scores moyens
#Группировка по полу с указанием средних оценок

sp.groupby('gender').mean()
```

```python
#Grouping by gender, race/ethnicity excluding indexes with mean reading score, mean math score. Renaming columns
#Groupement par genre, race/origine ethnique a l'exclusion des indices avec score moyen de lecture, score mathematique moyen. Renommer les colonnes
#Группировка по полу и расе/этничности без индексов с указанием средних оценок по математике и чтению. Переименование колонок.

sp.groupby(['gender', 'race/ethnicity'], as_index=False) \
    .aggregate({'reading_score': 'mean', 'math_score': 'mean'}) \
    .rename(columns = {'math_score': 'mean_math_score',
                      'reading_score': 'mean_reading_score'})
```

```python
#Grouping by gender, race/ethnicity with mean reading score, mean math score. Renaming columns
#Groupement par genre, race/origine ethnique avec score moyen de lecture, score mathematique moyen. Renommer les colonnes
#Группировка по полу и расе/этничности с указанием средних оценок по математике и чтению. Переименование колонок.

mss=sp.groupby(['gender', 'race/ethnicity']) \
    .aggregate({'reading_score': 'mean', 'math_score': 'mean'}) \
    .rename(columns = {'math_score': 'mean_math_score',
                      'reading_score': 'mean_reading_score'})
```

```python
#View of result
#Vue du resultat
#Представление результата

mss
```

```python
#View of indexes
#Vue des index
#Представление индексов

mss.index
```

```python
#Selecting attributes for male/group C and female/group B
#Selection des attributs pour homme/groupe C et femme/groupe B
#Выбор атрибутов для мужчин/группа С и женщин/группа B

mss.loc[[('male', 'group C'), ('female', 'group B')]]
```

```python
#View unique values for reading score
#Vue les valeurs uniques pour le score de lecture
#Представление уникальных значений среди оценок по чтению

sp.reading_score.unique()
```

```python
#Number unique values for writing score
#Nombre des valeurs uniques pour l'ecriture score
#Количество уникальных значений среди оценок по письму

sp.writing_score.nunique()
```

```python
#Number unique values for reading score in each categories
#Nombre des valeurs uniques pour le score de lecture dans chaque categorie
#Количество уникальных значений среди оценок по чтению в каждой категории

sp.groupby(['gender', 'race/ethnicity']).reading_score.nunique()
```

```python
#Sorting descending by reading score and gender
#Tri decroissant par score de lecture et genre
#Сортировка по убыванию оценки по чтению и по полу

sp.sort_values(['gender', 'reading_score'], ascending = False)
```

```python
#Top-10 results. Sorting descending by reading score and gender
#Top-10 resultats. Tri decroissant par score de lecture et genre
#Топ-10 результатов. Сортировка по убыванию оценки по чтению и по полу


sp.sort_values(['gender', 'reading_score'], ascending = False) \
    .groupby('gender').head(10)
```

```python
#Creating new attribute
#Creation d'un nouvel attribut
#Cоздание нового атрибута

sp['total_score'] = sp.math_score + sp.reading_score + sp.writing_score
```

```python
#View new attribute
#Vue du nouvel attribut
#Представление нового атрибута

sp.total_score
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#Calculating the natural logarithm of the total score
#Calculation du logarithme naturel du score total
#Вычисление натурального логарифма суммарной оценки

np.log(sp.total_score)
```

```python
#Creating new attribute
#Creation d'un nouvel attribut
#Cоздание нового атрибута

sp=sp.assign(total_score_log = np.log(sp.total_score))
```
