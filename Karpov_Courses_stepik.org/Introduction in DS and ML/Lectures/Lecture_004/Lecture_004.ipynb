```python
#Importing libraries for workflow
#Importer des bibliotheques pour le flux de travail
#Импорт библиотек для работ

import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading data frame into sp (Students Performance)
#Chargement de la dataframe dans sp
#Загрузка данных в переменную sp

sp=pd.read_csv('C:/Users/9/Downloads/StudentsPerformance.csv')
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#Renaming columns
#Renommer les colonnes
#Переименование колонок

sp = sp \
    .rename(columns = 
           {'parental level of education': 'parental_level_of_education',
            'test preparation course': 'test_preparation_course',
            'math score': 'math_score',
            'reading score': 'reading_score',
            'writing score': 'writing_score'
            })
```

```python
#Plotting histogram
#Trace de l'histogramme
#Построение гистограммы

sp.reading_score.hist()
```

```python
#Scatter plot
#Nuage de points
#Диаграмма рассеяния

sp.plot.scatter(x='math_score', y='reading_score')
```

```python
#Scatter plot with regression curve separated for gender
#Nuage de points avec courbe de regression separeee pout le genre
#Диаграмма рассеяния c линией регрессии по каждому полу

sns.lmplot(x='math_score', y='reading_score', hue='gender', data=sp)
```

```python
#Scatter plot without regression curve separated for gender. Renaming axis.
#Nuage de points sans courbe de regression separeee pout le genre. Renommer l'axe.
#Диаграмма рассеяния без линии регрессии по каждому полу. Подпись осей.

ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=sp, fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
