```python
#Importing libraries for workflow
#Importer des bibliotheques pour le flux de travail
#Импорт библиотек для работ

import pandas as pd
import numpy as np
```

```python
#Loading data frame into sp (Students Performance)
#Chargement de la dataframe dans sp
#Загрузка данных в переменную sp

sp=pd.read_csv('C:/Users/9/Downloads/StudentsPerformance.csv')
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#View of a series where the gender id female
#Vue d'une serie ou le genre est feminin
#Представление серии в которой пол - женский

sp.gender == 'female'
```

```python
#Selecting attributes gender and writing score where the gender id female
#Selection des attibuts de genre et score d'ecriture ou le genre est feminin
#Выбор атрибутов гендер и оценка по письму, где гендер - женский

sp.loc[sp.gender == 'female', ['gender', 'writing score']]
```

```python
#Creating variable mean writing score
#Creation d'un score d'ecriture moyen variable
#Создание переменной средней оценки по письму

mean_sp_ws = sp['writing score'].mean()
```

```python
#Selecting lines where writing score is above mean writing score
#Selectiom des lignes ou le score d'ecriture est superieur au score d'ecriture moyen
#Выбор строк в которой оценка по письму выше средней

sp.loc[sp['writing score'] > mean_sp_ws]
```

```python
#Creating a variable for a condition query
#Creation d'une variable pour une requete de condition
#Создание переменной для условия по запросу

query = (sp['writing score'] > mean_sp_ws) & (sp.gender == 'female')
```

```python
#Show the first 10 lines that match the query
#Afficher les 10 premieres lignes correpsopdant a la requete
#Показать 10 первых строк удовлетворяющих условию

sp.loc[query].head(10)
```

```python
#Renaming columns
#Renommer les colonnes
#Переименование колонок

sp = sp \
    .rename(columns = 
           {'parental level of education': 'parental_level_of_education',
            'test preparation course': 'test_preparation_course',
            'math score': 'math_score',
            'reading score': 'reading_score',
            'writing score': 'writing_score'
            })
```

```python
#Show the first 10 lines
#Afficher les 10 premieres lignes
#Показать 10 первых строк

sp.head(10)
```

```python
#Selecting lines where writing score is above 74
#Selectiom des lignes ou le score d'ecriture est superieur 74
#Выбор строк в которой оценка по письму выше 74

sp.query('writing_score > 74')
```

```python
#Selecting lines where reading score is above 90 and gender is male
#Selection des lignes ou le score de lecture est superieur 90 et genre est masculin
#Выбор строк в которой оценка по чтению выше 90 и гендер мужской

sp.query("gender == 'male' & reading_score > 90")
```

```python
#Creating a variable for a condition query
#Creation d'une variable pour une requete de condition
#Создание переменной для условия по запросу

reading_score_query = 95
```

```python
#View of query
#Vue de la requete
#Представление запроса

sp.query('reading_score > @reading_score_query')
```

```python
#Creating list
#Creation de liste
#Создание списка

sc = [i for i in list(sp) if 'score' in i]
```

```python
#View of list
#Vue de la liste
#Представление списка

sc
```

```python
#Show the first 10 lines from list
#Afficher les 10 premieres lignes de la liste
#Показать 10 первых строк из списка

sp[sc].head(10)
```

```python
#Using the filter function
#Utilisation de la fonction de filter
#Использование функции фильтр

sp.filter(like='score')
```

```python
#Using the filter function for columns
#Utilisation de la fonction de filter pour les colonnes
#Использование функции фильтр по столбцам

sp.filter(like='o', axis=1)
```
