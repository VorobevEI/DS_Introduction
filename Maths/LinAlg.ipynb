```python
k = 2
b = 1

def fun(x, y):
    return k*x**2 + b

x = np.arange(-3.0, 3.0, 0.05)
y = np.arange(-5.0, 5.0, 0.05)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
print('X.shape =', X.shape)
print('Y.shape =', Y.shape)
print('x.shape =', x.shape)
print('y.shape =', y.shape)
```

```python
np.ravel(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X_Label')
ax.set_ylabel('Y_Label')
ax.set_zlabel('Z_Label')

plt.show()
```

```python
gx, gy = np.gradient(Z)
```

```python
gx
```

```python
gy
```

```python
plt.streamplot(X, Y, gx, gy)
plt.show()
```

```python
X = np.array((0))
Y = np.array((0))
U = np.array((5))
V = np.array((8))

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V, units='xy', scale=1)

plt.grid()
ax.set_aspect('equal')
plt.xlim(0, 8)
plt.ylim(0, 10)

plt.title(' ')
plt.show()
```

```python
V = np.array([[5,8], [-6, -3]])
origin = np.array([[0,0], [0,0]])

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.quiver(*origin, V[:,0], V[:,1], color=['r','b'], scale = 21)
plt.grid()
plt.xlim(-7, 8)
plt.ylim(-5, 10)
plt.show()
```

```python
V = np.array([[5,8], [-6, -3]])
origin = np.array([[0, 0], [0, 0]])
v12 = V[0] + V[1]

plt.quiver(*origin, v12[0], v12[1])
plt.quiver(*origin, V[:, 0], V[:,1], color=['r', 'b'], scale = 21)
plt.grid()
plt.xlim(-10, 10)
plt.ylim(-5, 10)
plt.show()
```
